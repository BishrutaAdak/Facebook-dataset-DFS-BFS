```python
import networkx as netx
import matplotlib.pyplot as plot
```

```python
G = netx.read_edgelist('facebook_combined.txt')

print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
print(f"Is the graph directed: {G.is_directed()}")
print(f"Graph density: {netx.density(G)}")

start_node = '0'
goal_node = '100'

if start_node not in G or goal_node not in G:
    raise ValueError("Start or goal node not in graph")
```

```python
plot.figure(figsize=(12, 12))

pos = netx.spring_layout(G)

netx.draw(G, pos, with_labels=False, node_size=20, node_color="black", edge_color="gray")

plot.show()
```

```python
plot.figure(figsize=(12, 12))

pos = netx.spring_layout(G)

netx.draw_networkx_nodes(G, pos, node_size=50, node_color="black")

plot.axis('off')

plot.show()
```

```python
H = netx.Graph()
for u, v in netx.bfs_edges(G, source=start_node):
    if u != goal_node:
        H.add_edge(u, v)

print("Nodes visited while performing BFS from start node to goal node:")
print(H.nodes())
print("Edges traversed while performing BFS from start node to goal node:")
print(H.edges())

pos = netx.spring_layout(G)
netx.draw(H, pos, with_labels=True, node_color="lightblue", edge_color="gray")
plot.show()
```
